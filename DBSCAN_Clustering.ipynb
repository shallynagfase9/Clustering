#DBSCAN Clustering

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
%matplotlib inline


# DBSCAN clustering can be used to handle or cluster non-linear data points scattered in a efficient way
```

```python
X,y=make_moons(n_samples=250,noise=0.15)
```

```python
X
```

```python
y
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
##feature scaling(Standard Scaling)
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_scaled=scaler.fit_transform(X)
```

```python
dbcan=DBSCAN(eps=0.3)  # eps is radius
```

```python
dbcan.fit(X_scaled)
```

```python
dbcan.labels_  # -1 are my outliers
```

```python
plt.scatter(X[:,0],X[:,1],c=dbcan.labels_)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python

from sklearn.cluster import KMeans
```

```python
# Elbow method to select K-value
wcss=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(X_train)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python

## plot elbow curve
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.xlabel("Number of Clustrers")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans=KMeans(n_clusters=3,init="k-means++")
```

```python

y_labels=kmeans.fit_predict(X_train)
```

```python
plt.scatter(X_train[:,0],X_train[:,1],c=y_labels)
```

```python

## knee locator
# This will actually help you to find the K value
!pip install kneed
from kneed import KneeLocator
k1=KneeLocator(range(1,11),wcss,curve='convex',direction='decreasing')
k1.elbow
```

```python
## silhouette score
from sklearn.metrics import silhouette_score
```

```python

silhouette_coefficients=[]
for k in range(2,11):
    kmeans=KMeans(n_clusters=k,init="k-means++")
    kmeans.fit(X_train)
    score=silhouette_score(X_train,kmeans.labels_)
    silhouette_coefficients.append(score)
```

```python

silhouette_coefficients
```

```python
## plotting silhouette score
plt.plot(range(2,11),silhouette_coefficients)
plt.xticks(range(2,11))
plt.xlabel("Number of Cluters")
plt.ylabel("Silhoutte Coeffecient")
plt.show()
```
